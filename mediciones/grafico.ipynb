```python
import sqlite3
import matplotlib.pyplot as plt
import numpy as np

# Conexión a la base de datos SQLite
conn = sqlite3.connect('sensor.sqlite')  # Cambia el nombre del archivo si es diferente
cursor = conn.cursor()

# Consulta a la base de datos
cursor.execute("""SELECT timestamp, luz FROM LDRData 
                ORDER BY timestamp""")
data = cursor.fetchall()

# Cierra la conexión
conn.close()

# Procesar datos
timestamps = [row[0] for row in data]
luz_values = [row[1] for row in data]

# Convertir timestamps a índices si no son necesarios para el gráfico
x = np.arange(len(timestamps))  # Índices para el eje X
y = np.array(luz_values)       # Valores de luz

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b', label='Luz')
plt.xlabel('Mediciones (Índice)')
plt.ylabel('Intensidad de Luz')
plt.title('Gráfico de Intensidad de Luz')
plt.legend()
plt.grid(True)
plt.show()
```
